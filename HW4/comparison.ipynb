```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
DCGAN = np.load('Results/FID_score/DC_FID.npy')
WGANGP = np.load('Results/FID_score/WGANGP_FID.npy')
ACGAN = np.load('Results/FID_score/ACGAN_FID.npy')
```

```python
plt.figure(figsize=(10,5))
plt.title("FID Scores comparsions for DCGAN vs WGAN vs ACGAN")
plt.plot(DCGAN,label="DCGAN",color='red')
plt.plot(WGANGP,label="WGAN",color='green')
plt.plot(ACGAN,label="ACGAN")
plt.xlabel("Epochs")
plt.ylabel("FID")
plt.legend()
plt.savefig('Results/FID_summary.jpg',
            format='jpeg',
            dpi=100,
            bbox_inches='tight')
plt.show()
```
