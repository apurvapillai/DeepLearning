```python
import numpy as np
import torch

def simulate_non_linear_function(x):
    return np.sign(np.sin(5 * np.pi * x))

x = np.linspace(-10, 10, 100)
y = simulate_non_linear_function(x)

# Convert to PyTorch tensors
x_tensor = torch.from_numpy(x).float().view(-1, 1)
y_tensor = torch.from_numpy(y).float().view(-1, 1)
```

```python
import torch.nn as nn

class DNNModel1(nn.Module):
    def __init__(self):
        super(DNNModel1, self).__init__()
        self.fc1 = nn.Linear(1, 64)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(64, 64)
        self.fc3 = nn.Linear(64, 1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

class DNNModel2(nn.Module):
    def __init__(self):
        super(DNNModel2, self).__init__()
        self.fc1 = nn.Linear(1, 128)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(128, 128)
        self.fc3 = nn.Linear(128, 1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

class DNNModel3(nn.Module):
    def __init__(self):
        super(DNNModel3, self).__init__()
        self.fc1 = nn.Linear(1, 256)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(256, 256)
        self.fc3 = nn.Linear(256, 1)

    def forward(self, x):
        x = self.relu(self.fc1(x))
        x = self.relu(self.fc2(x))
        x = self.fc3(x)
        return x

model1 = DNNModel1()
model2 = DNNModel2()
model3 = DNNModel3()
```

```python
criterion = nn.L1Loss()
optimizer1 = torch.optim.Adam(model1.parameters(), lr=0.001)
optimizer2 = torch.optim.Adam(model2.parameters(), lr=0.001)
optimizer3 = torch.optim.Adam(model3.parameters(), lr=0.001)

for epoch in range(20000):
    optimizer1.zero_grad()
    optimizer2.zero_grad()
    optimizer3.zero_grad()

    output1 = model1(x_tensor)
    output2 = model2(x_tensor)
    output3 = model3(x_tensor)

    loss1 = criterion(output1, y_tensor)
    loss2 = criterion(output2, y_tensor)
    loss3 = criterion(output3, y_tensor)

    loss1.backward()
    loss2.backward()
    loss3.backward()

    optimizer1.step()
    optimizer2.step()
    optimizer3.step()

    print(f'Epoch {epoch+1}, Loss1: {loss1.item():.4f}, Loss2: {loss2.item():.4f}, Loss3: {loss3.item():.4f}')
```

```python
import matplotlib.pyplot as plt

loss1_values = []
loss2_values = []
loss3_values = []

for epoch in range(20000):
    optimizer1.zero_grad()
    optimizer2.zero_grad()
    optimizer3.zero_grad()

    output1 = model1(x_tensor)
    output2 = model2(x_tensor)
    output3 = model3(x_tensor)

    loss1 = criterion(output1, y_tensor)
    loss2 = criterion(output2, y_tensor)
    loss3 = criterion(output3, y_tensor)

    loss1.backward()
    loss2.backward()
    loss3.backward()

    optimizer1.step()
    optimizer2.step()
    optimizer3.step()

    loss1_values.append(loss1.item())
    loss2_values.append(loss2.item())
    loss3_values.append(loss3.item())

    print(f'Epoch {epoch+1}, Loss1: {loss1.item():.4f}, Loss2: {loss2.item():.4f}, Loss3: {loss3.item():.4f}')

plt.plot(loss1_values, label='Model 1')
plt.plot(loss2_values, label='Model 2')
plt.plot(loss3_values, label='Model 3')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
y_pred1 = model1(x_tensor).detach().numpy()
y_pred2 = model2(x_tensor).detach().numpy()
y_pred3 = model3(x_tensor).detach().numpy()

plt.plot(x, y, label='Ground Truth')
plt.plot(x, y_pred1, label='Model 1')
plt.plot(x, y_pred2, label='Model 2')
plt.plot(x, y_pred3, label='Model 3')
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend(loc='upper right')
plt.show()
```
